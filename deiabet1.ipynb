```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
```

```python
#بارگذاری دیتا ست diabetes
diabet=load_diabetes()
data=pd.DataFrame(diabet.data ,columns=diabet.feature_names)
data["target"]=diabet.target
data.head()
```

```python
data.shape
```

```python
#روش اول Kmean
#پیدا کردن kمناسب با استفده از elbow

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from seaborn import set_style
set_style('darkgrid')


wcss = []  # مجموع مربعات درون کلاسترها
max_k = 8 # حداکثر تعداد کلاسترها

for k in range(1, max_k + 1):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)  # ذخیره مجموع مربعات درون کلاسترها

# رسم نمودار روش آرنج
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_k + 1), wcss, marker='o')
plt.title('elbow')
plt.xlabel('the number of clusters (k)')
plt.ylabel('WCSS')
plt.show()



# محاسبه نقطه عطف  به صورت خودکار
diff = np.diff(wcss)
diff2 = np.diff(diff)

# پیدا کردن نقطه عطف
optimal_k = np.argmin(diff2) + 2  # +2 برای تنظیم ایندکس به تعداد کلاستر


# چاپ تعداد کلاستر بهینه
print(f"the optimal number of clustters: {optimal_k}")


#  K-Means با تعداد کلاسترهای بهینه
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
kmeans.fit(data)

# اضافه کردن برچسب کلاستر به داده‌ها
data['Cluster'] = kmeans.labels_

```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error
from collections import Counter


target_col = 'target'  # ستون هدف (مقادیر عددی)
cluster_col = 'Cluster'  # ستون کلاستر (مقادیر دسته‌ای)

# جدا کردن ویژگی‌ها، برچسب‌ها و کلاسترها
X = data.drop(columns=[target_col, cluster_col]).values
y = data[target_col].values
clusters = data[cluster_col].values

# تقسیم داده به train و test
X_train, X_test, y_train, y_test, clusters_train, clusters_test = train_test_split( X, y, clusters, test_size=0.13, random_state=0
)

# نرمال‌سازی داده‌ها
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# تعیین k برای KNN (جذر تعداد نمونه‌های آموزش)
k = int(np.sqrt(len(y_train)))

# لیست‌های ذخیره‌سازی نتایج
final_predictions = []  # پیش‌بینی نهایی برای هر نمونه تست
feature_predictions_all = []  # پیش‌بینی‌های هر ویژگی برای همه نمونه‌های تست

# پردازش هر نمونه تست
for i in range(len(X_test)):
    feature_predictions = []  # پیش‌بینی‌های هر ویژگی برای این نمونه تست

    # پردازش هر ویژگی به صورت جداگانه
    for feature_idx in range(X.shape[1]):
        # انتخاب فقط یک ویژگی از داده‌های آموزش
        X_train_feature = X_train_scaled[:, feature_idx].reshape(-1, 1)
        X_test_feature = X_test_scaled[i, feature_idx].reshape(1, -1)

        # آموزش KNN برای کلاسترینگ روی این ویژگی
        knn_cluster = KNeighborsClassifier(n_neighbors=k)
        knn_cluster.fit(X_train_feature, clusters_train)

        # پیش‌بینی کلاستر برای این ویژگی
        predicted_cluster = knn_cluster.predict(X_test_feature)[0]

        # یافتن همسایگان با کلاستر پیش‌بینی شده در این ویژگی
        neighbor_mask = (clusters_train == predicted_cluster)
        y_neighbors = y_train[neighbor_mask]

        # محاسبه میانگین تارگت‌های همسایگان این ویژگی
        if len(y_neighbors) > 0:
            feature_pred = np.mean(y_neighbors)
        else:
            feature_pred = np.mean(y_train)  # اگر همسایه‌ای نبود

        feature_predictions.append(feature_pred)

    # ذخیره پیش‌بینی‌های تمام ویژگی‌ها برای این نمونه تست
    feature_predictions_all.append(feature_predictions)

    # محاسبه پیش‌بینی نهایی (میانگین پیش‌بینی‌های تمام ویژگی‌ها)
    final_prediction = np.mean(feature_predictions)
    final_predictions.append(final_prediction)

# محاسبه MSE
mse = mean_squared_error(y_test, final_predictions)

# نمایش نتایج
results = pd.DataFrame({
    'True Target': y_test,
    'Final Prediction': final_predictions,
})

# اضافه کردن پیش‌بینی‌های هر ویژگی به نتایج
for feature_idx in range(X.shape[1]):
    results[f'Feature {feature_idx} Pred'] = [preds[feature_idx] for preds in feature_predictions_all]

print("="*50)
print("نمونه‌ای از پیش‌بینی‌ها:")
print(results.head())
print("="*50)
print(f"MSE: {mse:.4f}")
print("="*50)
```

روش فیچر به فیچر و در نهایت استفاده از رگرسیون خطی برای همسایگان مجوریتی

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from collections import Counter

# فرض می‌کنیم داده‌ها در متغیر data موجود است
target_col = 'target'  # ستون هدف (مقادیر عددی)
cluster_col = 'Cluster'  # ستون کلاستر (مقادیر دسته‌ای)

# جدا کردن ویژگی‌ها، برچسب‌ها و کلاسترها
X = data.drop(columns=[target_col, cluster_col]).values
y = data[target_col].values
clusters = data[cluster_col].values

# تقسیم داده به train و test
X_train, X_test, y_train, y_test, clusters_train, clusters_test = train_test_split(
    X, y, clusters, test_size=0.13, random_state=0
)

# نرمال‌سازی داده‌ها
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# تعیین k برای KNN (جذر تعداد نمونه‌های آموزش)
k = int(np.sqrt(len(y_train)))

# لیست‌های ذخیره‌سازی نتایج
final_cluster_predictions = []  # کلاستر غالب برای هر نمونه تست
linear_regression_predictions = []  # پیش‌بینی رگرسیون خطی روی همسایگان

# پردازش هر نمونه تست
for i in range(len(X_test)):
    feature_cluster_predictions = []

    # پردازش هر ویژگی به صورت جداگانه
    for feature_idx in range(X.shape[1]):
        X_train_feature = X_train_scaled[:, feature_idx].reshape(-1, 1)
        X_test_feature = X_test_scaled[i, feature_idx].reshape(1, -1)

        knn_cluster = KNeighborsClassifier(n_neighbors=k)
        knn_cluster.fit(X_train_feature, clusters_train)
        predicted_cluster = knn_cluster.predict(X_test_feature)[0]
        feature_cluster_predictions.append(predicted_cluster)

    # کلاستر غالب (مد) بین تمام ویژگی‌ها
    final_cluster = Counter(feature_cluster_predictions).most_common(1)[0][0]
    final_cluster_predictions.append(final_cluster)

    # یافتن همسایگان در فضای تمام ویژگی‌ها که کلاسترشان = کلاستر غالب است
    neighbor_mask = (clusters_train == final_cluster)
    X_train_filtered = X_train_scaled[neighbor_mask]
    y_train_filtered = y_train[neighbor_mask]

    if len(y_train_filtered) == 0:
        # اگر هیچ همسایه‌ای با کلاستر غالب نبود، از میانگین کل داده‌ها استفاده کن
        linear_pred = np.mean(y_train)
    else:
        # رگرسیون خطی روی همسایگان با کلاستر غالب
        linear_reg = LinearRegression()
        linear_reg.fit(X_train_filtered, y_train_filtered)
        linear_pred = linear_reg.predict(X_test_scaled[i].reshape(1, -1))[0]

    # ذخیره نتایج
    linear_regression_predictions.append(linear_pred)

# محاسبه MSE برای رگرسیون خطی
mse_linear = mean_squared_error(y_test, linear_regression_predictions)

# نمایش نتایج
results = pd.DataFrame({
    'True Target': y_test,
    'Final Cluster': final_cluster_predictions,
    'Linear Regression Prediction': linear_regression_predictions
})

print("="*50)
print("نمونه‌ای از پیش‌بینی‌ها:")
print(results.head())
print("="*50)
print(f"میانگین مربعات خطا (MSE) برای رگرسیون خطی: {mse_linear:.4f}")
print("="*50)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# فرض کنید داده‌ها در متغیر `data` موجود است
target_col = 'target'  # ستون هدف (مقادیر عددی)
feature_cols = data.columns.drop([target_col])  # لیست ستون‌های ویژگی

# جدا کردن ویژگی‌ها و تارگت
X = data[feature_cols].values
y = data[target_col].values

# تقسیم داده به train و test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.13, random_state=0)

# نرمال‌سازی داده‌ها
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# لیست ذخیره پیش‌بینی‌های هر ویژگی برای هر نمونه تست
feature_predictions_per_test = []

# پردازش هر نمونه تست
for i in range(len(X_test)):
    # لیست ذخیره پیش‌بینی‌های هر ویژگی برای این نمونه تست
    single_test_feature_predictions = []

    # پردازش هر ویژگی به صورت جداگانه
    for feature_idx in range(X.shape[1]):
        # انتخاب فقط یک ویژگی از داده‌های آموزش
        X_train_feature = X_train_scaled[:, feature_idx].reshape(-1, 1)
        y_train_target = y_train

        # آموزش رگرسیون خطی روی این ویژگی
        lr = LinearRegression()
        lr.fit(X_train_feature, y_train_target)

        # پیش‌بینی برای این ویژگی در نمونه تست فعلی
        X_test_feature = X_test_scaled[i, feature_idx].reshape(1, -1)
        pred = lr.predict(X_test_feature)[0]
        single_test_feature_predictions.append(pred)

    # ذخیره بردار پیش‌بینی‌های این نمونه تست
    feature_predictions_per_test.append(single_test_feature_predictions)

# تبدیل لیست به آرایه numpy
feature_predictions_per_test = np.array(feature_predictions_per_test)

# حالا رگرسیون خطی روی بردار پیش‌بینی‌های هر نمونه تست
final_predictions = []
lr_final = LinearRegression()

# آموزش مدل نهایی روی بردار پیش‌بینی‌های ویژگی‌ها
lr_final.fit(feature_predictions_per_test, y_test)

# پیش‌بینی نهایی
final_predictions = lr_final.predict(feature_predictions_per_test)

# محاسبه MSE برای دو روش:
# 1. MSE برای میانگین پیش‌بینی‌های هر ویژگی
mean_feature_predictions = np.mean(feature_predictions_per_test, axis=1)
mse_mean_features = mean_squared_error(y_test, mean_feature_predictions)

# 2. MSE برای رگرسیون خطی روی بردار پیش‌بینی‌ها
mse_final_lr = mean_squared_error(y_test, final_predictions)

# نمایش نتایج
print("="*50)
print("MSE ها:")
print(f"- MSE (میانگین پیش‌بینی‌های هر ویژگی): {mse_mean_features:.4f}")
print(f"- MSE (رگرسیون خطی روی بردار پیش‌بینی‌ها): {mse_final_lr:.4f}")
print("="*50)

# نمایش چند نمونه پیش‌بینی
results = pd.DataFrame({
    'True Target': y_test,
    'Mean Feature Predictions': mean_feature_predictions,
    'Final LR Predictions': final_predictions
})
print(results.head(10))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# فرض کنید داده‌ها در متغیر `data` موجود است
target_col = 'target'  # ستون هدف (عددی)
feature_cols = data.columns.drop(target_col)  # لیست ویژگی‌ها

# آماده‌سازی داده‌ها
X = data[feature_cols].values
y = data[target_col].values

# تقسیم داده به آموزش و آزمون
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.13, random_state=0)

# نرمال‌سازی داده‌ها
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# تعیین تعداد همسایه‌ها (k)
k = int(np.sqrt(len(X_train)))

# لیست ذخیره پیش‌بینی‌های هر ویژگی برای هر نمونه تست
feature_preds = []

# 1. پیش‌بینی هر ویژگی به صورت جداگانه با KNN رگرسیون
for i in range(X_test.shape[0]):
    sample_preds = []
    for feature_idx in range(X.shape[1]):
        # آموزش مدل روی یک ویژگی
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(X_train_scaled[:, feature_idx].reshape(-1, 1), y_train)

        # پیش‌بینی برای ویژگی فعلی
        pred = knn.predict(X_test_scaled[i, feature_idx].reshape(1, -1))[0]
        sample_preds.append(pred)
    feature_preds.append(sample_preds)

feature_preds = np.array(feature_preds)

# 2. پیش‌بینی نهایی با KNN رگرسیون روی بردار ویژگی‌ها
final_knn = KNeighborsRegressor(n_neighbors=k)
final_knn.fit(feature_preds, y_test)
final_preds = final_knn.predict(feature_preds)

# محاسبه MSE
mse_individual = mean_squared_error(y_test, np.mean(feature_preds, axis=1))
mse_final = mean_squared_error(y_test, final_preds)

print("MSE (میانگین پیش‌بینی‌های تکی):", mse_individual)
print("MSE (پیش‌بینی نهایی با KNN):", mse_final)

# نمایش نتایج
results = pd.DataFrame({
    'واقعی': y_test,
    'میانگین پیش‌بینی‌ها': np.mean(feature_preds, axis=1),
    'پیش‌بینی نهایی KNN': final_preds
})
print(results.head(10))
```
